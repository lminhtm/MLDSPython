## Thao tác cột trong DF

```python
# Thêm cột: cột trong df là 1 Series
df['class'] = pd.Series(['DS1','DS2','DS3','DS4'],index=['a','b','c','d'])
df
```

```python
# tạo cột mới từ cột đã có trong df
df['Type'] = df['age'].map(lambda x: 1 if x>=20 else 0)
df
```

```python
df['tmp'] = df.age + 10
df
```

```python
# xóa cột
df.drop('tmp',axis=1)
```

```python
# đổi tên cột
df.rename(columns={'name':'ten','age':'tuoi'}) # ko lưu
```
