## Chapter 2 - Exercise5: Đọc tập tin data.txt là chiều cao của 10 sinh viên
- In ra chiều cao cao nhất, chiều cao thấp nhất
- Tính và in ra chiều cao trung bình
- Tính và in ra độ lệch chuẩn (standard deviation) theo công thức:


```python
import math
f = open("data.txt")
content = f.read()
chieu_cao = [float(x) for x in content.split(",")]
print(chieu_cao)

print("Chieu cao cao nhat:", max(chieu_cao))
print("Chieu cao thap nhat:", min(chieu_cao))
avg = sum(chieu_cao)/len(chieu_cao)
print("Chieu cao trung binh:", avg)

N = len(chieu_cao)
sum = 0
for i in range(0, 7):
    sum += (chieu_cao[i] - avg) ** 2
do_lech_chuan = math.sqrt(sum/N)
print("Do lech chuan:", do_lech_chuan)

f.close()
```
