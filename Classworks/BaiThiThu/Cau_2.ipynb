```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv("data/spam.csv")
df.head()
```

```python
stopwords = set(STOPWORDS)
```

```python
spam_words =' '.join(df['content'])
```

```python
# Vẽ WordClouds cho dữ liệu này
wc = WordCloud(
    background_color='white',
    stopwords=stopwords
)
# generate the word cloud
wc.generate(spam_words)
# display the word clouds
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
```

```python
vr = np.array(Image.open('data/virus.png'))
plt.imshow(vr, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Vẽ wordclouds với vr_mask
wc1 = WordCloud(background_color='white', max_words=1000,
mask=vr, stopwords=stopwords)
wc1.generate(spam_words)
plt.figure(figsize=(10, 12))
plt.imshow(wc1, interpolation='bilinear')
plt.axis('off')
plt.show()
```
