# Câu 2. WordCloud

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

### 1. Đọc dữ liệu và tạo đoạn text từ cột content. Sau đó thực hiện chuẩn hóa đoạn text (loại bỏ các từ không quan trọng)

```python
df = pd.read_csv('data/tech.csv')
df.head()
```

```python
# Build words
words = ' '.join(df['Content'])
```

```python
# Build stop words
stopwords = set(STOPWORDS)
list_of_words = ['The', 'people', 'U', 'will', 'one', 'much', 'many']
stopwords.update(list_of_words)
```

### 2. Tạo biểu đồ Wordcloud

```python
# Create WordCloud
wc = WordCloud(background_color='white', stopwords=stopwords)
wc.generate(words)

# Draw
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 3. Tạo biểu đồ Wordcloud with mask

```python
# Create WordCloud with mask
wc_mask = np.array(Image.open('data/bird.jpg'))
wc = WordCloud(background_color='white', mask=wc_mask, stopwords=stopwords)
wc.generate(words)

# Draw
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
