# Word Cloud

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import warnings
warnings.filterwarnings('ignore')
```

## Prepare data

```python
# Load data
df = pd.read_csv("data/Youtube04-Eminem.csv", encoding ="latin-1") 
df.head()
```

```python
# Build words and stop words
comment_words = ' '.join(df['CONTENT'])
stopwords = set(STOPWORDS)
```

## Draw (matplotlib)

```python
# Create WordCloud
wc = WordCloud(background_color='black', stopwords=stopwords)
wc.generate(comment_words)

# Draw
plt.figure(figsize=(10, 12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Draw with mask

```python
# Create WordCloud with mask
wc_mask = np.array(Image.open('data/heart.jpg'))
wc = WordCloud(background_color='white', max_words=1000, mask=wc_mask, stopwords=stopwords)
wc.generate(comment_words)

# Draw
plt.figure(figsize=(8, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
