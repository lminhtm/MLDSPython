```python
# kiểm tra Null
df.isnull().sum()
```

```python
# kiểm tra duplicate
df.duplicated().any()
```

```python
# Câu 2: Cho biết thu nhập bình quân đầu người và tuổi thọ trung bình 
# của item đầu (làm tròn 2 số lẻ)
print('Thu nhập bình quân đầu người của item đầu là:', round(df.iloc[0,0],2))
print('Tuổi thọ trung bình của item đầu là:', round(df.iloc[0,1],2))
```

```python
# Câu 2: Cho biết thu nhập bình quân đầu người và tuổi thọ trung bình 
# của item cuối
print('Thu nhập bình quân đầu người của item đầu là:', round(df.iloc[-1,0],2))
print('Tuổi thọ trung bình của item đầu là:', round(df.iloc[-1,1],2))
```

```python
# Câu 3: Thử vẽ biểu đồ line liên hệ giữa gdp_cap và life_exp 
# với x-axis: gdp_cap, y-axis: life_exp. 

# Biểu đồ này có thể xem được không? 
# Nếu không thì bạn hãy đề xuất một loại biểu đồ phù hợp?
plt.plot(df['gdp_cap'], df['life_exp'])
plt.show()
```

```python
# Có thể thay biểu đồ line thành biểu đồ scatter plot
plt.scatter(df['gdp_cap'], df['life_exp'])
plt.show()
```

```python
# Câu 4: Vẽ biểu đồ histogram của life_exp, màu cột xanh, viền đỏ (mặc định là 10 bins)
# Bạn nhận xét gì qua biểu đồ vừa vẽ
plt.hist(df['life_exp'], color='g', edgecolor='r')
plt.show()
```

```python
# Câu 6: Tạo scatter plot của gdp_gap và life_exp nhưng sử dụng plt.xscale('log'). 
# Khi trực quan hóa dữ liệu thay đổi trong phạm vi rất rộng, 
# thang đo logarit plt.xscale('log') cho phép 
# chúng ta hình dung các thay đổi một cách trực quan hơn.
plt.scatter(df['gdp_cap'], df['life_exp'])
plt.xscale('log')
# Thiết lập xlabel, ylabel, title
plt.xlabel('GDP per Capital [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.show()
```

```python
# Câu 7: Tạo Scatter plot của gdp_gap và life_exp, sử dụng plt.xscale('log'). 
# Thiết lập xlabel, ylabel, title
# Với: tick_val = [1000,10000,100000] và 
# tick_lab = ['1k','10k','100k'] => plt.xticks(tick_val, tick_lab)
plt.scatter(df['gdp_cap'], df['life_exp'])
plt.xscale('log')
plt.xlabel('GDP per Capital [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']
plt.xticks(tick_val, tick_lab)
plt.show()
```

```python
# Câu 8: Vẽ scatter plot của gdp_cap và life_exp, với s = pop * 2, màu magenta
plt.scatter(df['gdp_cap'], df['life_exp'] , s = df['pop'] * 2, color='m')
# Thiết lập xlabel, ylabel, title và
# plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.xscale('log')
plt.xlabel('GDP per Capital [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.show()
```

```python
# Câu 9: Vẽ scatter plot của gdp_cap và life_exp, với s = pop*2, 
# màu c = color (giá trị color trong file dữ liệu) , alpha=0.8
plt.scatter(x = df['gdp_cap'], y = df['life_exp'], s = df['pop'] * 2,
c= df['color'], alpha=0.8)
# Thiết lập xlabel, ylabel, title
# và plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.xscale('log')
plt.xlabel('GDP per Capital [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])
plt.show()
# Bạn nhận xét gì về biểu đồ vừa vẽ
```

```python
# Câu 10: Vẽ scatter plot của gdp_cap, life_exp, 
# với s = pop*2, màu c = color, alpha=0.8


plt.scatter(x = df['gdp_cap'], y = df['life_exp'],
s = df['pop'] * 2, c= df['color'], alpha=0.8)
# Thiết lập xlabel, ylabel, title và
# plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
plt.xscale('log')
plt.xlabel('GDP per Capital [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])
# Thêm text cho 2 nơi là India và China:
# plt.text(1550, 71, 'India'), plt.text(5700, 80, 'China')
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')
# Thêm lưới cho biểu đồ
plt.grid()
plt.show()
```
