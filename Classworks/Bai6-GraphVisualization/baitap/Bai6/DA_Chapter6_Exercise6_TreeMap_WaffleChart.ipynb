## Chapter 6 - Exercise 6: TreeMap - Waffle Chart

#### Thực hành vẽ TreeMap và Waffle Chart trên cùng tập dữ liệu

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
import squarify 
```

### Part 1: TreeMap

```python
# Cho dữ liệu: 
df = pd.DataFrame(
    {
        'Name': ['Hillary Clinton', 'Donald Trump', 'Others'],
        'Virginia': [1981473, 1769443, 233715],
        'Maryland': [1677928, 943169, 160349],
        'West Virginia': [188794, 489371, 36258],
    }
)
```

```python
# In nội dung của df
df
```

## Câu 1: 
- Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Virginia

```python
# Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Virginia
plt.figure(figsize=(8,6))
squarify.plot(sizes=df.Virginia.values, label=df.Name.values)
plt.title("2016 Virginia Presidential Election Results", fontsize=18, color = 'b')
plt.axis('off')
plt.show()
```

## Câu 2:
- Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Maryland
- Vẽ cho 3 màu '#3c548c', '#7b04ab', '#ab2493'

```python
plt.figure(figsize=(8,6))
some_colors = ['#3c548c', '#7b04ab', '#ab2493']
squarify.plot(sizes=df.Maryland.values, label=df.Name.values, color=some_colors,
pad=True, alpha=0.6)
plt.title("2016 Maryland Presidential Election Results", fontsize=18, color = 'b')
plt.axis('off')
plt.show()
```

## Câu 3:
- Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở West Virginia
- Vẽ cho 3 màu ngẫu nhiên

```python
np.random.seed(42)

plt.figure(figsize=(8,6))
cmap = cm.mpl.cm.get_cmap()
squarify.plot(sizes=df['West Virginia'].values, label=df.Name.values, pad=True,
alpha=0.6,
color=[cmap(np.random.random()) for i in range(len(df))])
plt.title("2016 West Virginia Presidential Election Results", fontsize=18, color = 'b')
plt.axis('off')
plt.show()
```

## Câu 4:
- Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở West Virginia
- Vẽ cho 3 màu ngẫu nhiên và có số liệu

```python
np.random.seed(42)
labels = [df.loc[i,'Name'] + '\n' + '(' + '{:,}'.format(df.loc[i,'West Virginia']) + ')' for i in range(len(df))]
plt.figure(figsize=(8,6))
cmap = cm.mpl.cm.get_cmap()
squarify.plot(sizes=df['West Virginia'].values, label=labels, pad=True,
alpha=0.6,
color=[cmap(np.random.random()) for i in range(len(df))])
plt.title("2016 West Virginia Presidential Election Results", fontsize=18, color = 'b')
plt.axis('off')
plt.show()
```

## Câu 2:
- Vẽ Waffle Chart , lưu ý chia lại Tỷ lệ tổng giá trị là 1/100000

```python
total = round(df.Total/100000)
total.values
```

```python
# Vẽ Waffle Chart
# Vẽ Waffle Chart
fig = plt.figure(
FigureClass=Waffle,
rows=df.shape[0],
values=list(total),
labels=list(df.Name),
figsize=(10, 5),
legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)
}
)
plt.title("2016 Presidential Election Results", fontsize=18, color = 'b')
plt.show()
```
