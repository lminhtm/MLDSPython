# Swarm Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

## Chuẩn bị dữ liệu

```python
# Đọc tập tin dữ liệu , dữ liệu về các vận động viên đã tham gia Thế vận hội Rio 2016
df = pd.read_csv('data/athletes.csv', index_col=0)
df.head()
```

```python
# 300 VĐV (chọn ngẫu nhiên) 
np.random.seed(63)
df_300 = df.sample(n=300)
df_300.head()
```

## Seaborn

```python
sns.swarmplot(data=df_300, x='sex', y='weight')

plt.xlabel('Giới tính')
plt.ylabel('Cân nặng (kg)')
plt.show()
```

```python
df_ = df[df['sport'].isin(['volleyball','basketball'])]
sns.swarmplot(data=df_, x='sport', y='height', hue='sex')

plt.title('Biểu đồ Swarmplot chiều cao theo môn thi đấu')
plt.xlabel('Môn thi đấu')
plt.ylabel('Chiều cao')
plt.show()
```

```python
# Vẽ stripplot
sns.stripplot(data=df_300, x='sex', y='weight')

plt.xlabel('Giới tính')
plt.ylabel('Cân nặng')
plt.show()
```

```python
sns.catplot(x="sex", y="height", col="sport", data=df_, kind = 'swarm');
```
