# Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Chuẩn bị dữ liệu

```python
# Đọc tập tin dữ liệu , dữ liệu về các vận động viên đã tham gia Thế vận hội Rio 2016
df = pd.read_csv('data/data.csv', index_col=0)
df.head()
```

```python
museum= pd.read_csv('data/museum_visitors.csv', index_col=0, parse_dates=True)
museum.head()
```

## Matplotlib

```python
plt.plot(df['year'], df['pop'], '--c')

plt.title('Dân số thế giới qua các năm ')
plt.xlabel('Năm')
plt.ylabel('Dân số (tỷ người)') 
plt.grid()
plt.show()
```

```python
plt.plot(museum.index, museum['Avila Adobe'], label='Avila Adobe')
plt.plot(museum.index, museum['Firehouse Museum'], label='Firehouse Museum')
plt.plot(museum.index, museum['Chinese American Museum'], label='Chinese American Museum')
plt.plot(museum.index, museum['America Tropical Interpretive Center'], label='America Tropical Interpretive Center')

plt.legend()
plt.grid()
plt.show()
```

## Seaborn

```python
sns.lineplot(data=museum)

plt.grid()
plt.show()
```

```python
sns.lineplot(data=museum,x=museum.index,y='America Tropical Interpretive Center')
sns.lineplot(data=museum,x=museum.index,y='Chinese American Museum')

plt.grid()
plt.show()
```
