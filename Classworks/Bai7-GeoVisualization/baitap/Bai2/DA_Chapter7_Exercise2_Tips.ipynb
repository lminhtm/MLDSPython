## Chapter 7 - Exercise 2: Tips

#### Cho dữ liệu tips có sẵn trong seaborn library. Hãy vẽ những biểu đồ theo yêu cầu:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dữ liệu tips có sẵn trong seaborn library
#total_bill: Total bill (cost of the meal), including tax, in US dollars
#tip: Tip (gratuity) in US dollars
#sex: Sex of person paying for the meal (0=male, 1=female)
#smoker: Smoker in party? (0=No, 1=Yes)
#day: Thur, Fri, Sat, Sun
#time: Dinner, Lunch
#size: Size of the party
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
```

```python
tips.head()
```

## Câu 1:

```python
# Câu 1: 
# a) Vẽ biểu đồ barplot với x là day, y là tổng các total_bill
# chia nhóm hue theo sex, alpha 0.7, ci='sd'
# cho nhận xét
sns.barplot(data=tips, x='day', y='total_bill', ci='sd', estimator=sum,
            hue='sex', alpha=0.7)
plt.show()
```
