## Chapter 7 - Exercise 3: Titanic

#### Cho dữ liệu titanic có sẵn trong seaborn library. Hãy vẽ những biểu đồ theo yêu cầu, và cho biết nhận xét sau biểu đồ vừa vẽ:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Load dữ liệu titanic có sẵn trong seaborn library
titanic = sns.load_dataset("titanic")
titanic.info()
titanic.head()
```

## Câu 1:

```python
# a) Vẽ stripplot thể hiện sự phân bố của fare theo class
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(8,6))
sns.stripplot(data=titanic, x='fare', y='class', jitter=True)
plt.show()
```

```python
# b) Vẽ stripplot thể hiện sự phân bố của fare theo class, phân nhóm theo sex
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(8,6))
sns.stripplot(data=titanic, x='fare', y='class', jitter=True, hue='sex')
plt.show()
```

## Câu 2:

```python
# Vẽ violinplot thể hiện sự phân bố của fare theo class
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(8,6))
sns.violinplot(data=titanic, x='fare', y='class', palette='husl')
plt.show()
```

## Câu 3:

```python
# Vẽ countplot đếm số lượng alive/not alive theo từng class
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(8,6))
sns.countplot(data=titanic, y="class", hue="alive")
plt.grid(True)
plt.show()
```

## Câu 4:

```python
# Vẽ pointplot thể hiện khả năng sống sót 'survived' theo class
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(8,6))
sns.pointplot(data=titanic, y='survived', x='class', capsize=.1)
plt.show()
```

## Câu 5:

## Câu 6:

```python
# Vẽ correlation matrix (heatmap) của titanic
# Bạn nhận xét gì về biểu đồ vừa tạo
plt.figure(figsize=(10,8))
sns.heatmap(titanic.corr())
plt.show()
```

## Câu 7:

```python
# Vẽ biểu đồ histogram của cột vé (Fare), 
# bổ sung các thông tin xlabel, ylabel, title
# Bạn nhận xét gì về biểu đồ vừa vẽ
binsVal = np.arange(0,600,30)
sns.displot(data=titanic, x='fare', bins=binsVal, kde=True,
height=5, aspect=1.5)
# Thiết lập các thông tin xlabel, ylabel, title
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histogram')
plt.show()
```
