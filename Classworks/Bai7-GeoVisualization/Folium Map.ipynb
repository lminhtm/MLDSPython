# Folium Map

```python
import pandas as pd
import numpy as np
import folium 
import matplotlib.pyplot as plt
```

## 1. Create Map

```python
# World map
world_map = folium.Map()
world_map
```

```python
# vn map
vn_map = folium.Map(location=[10.762622, 106.660172], zoom_start=8, tiles='Stamen Terrain')
vn_map
```

```python
# Custom map
hcm_latitude = 10.762622 
hcm_longitude = 106.660172

hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
folium.TileLayer('Stamen Terrain').add_to(hcm_map)
folium.TileLayer('Stamen Toner').add_to(hcm_map)
folium.TileLayer('Stamen Water Color').add_to(hcm_map)
folium.TileLayer('cartodbpositron').add_to(hcm_map)
folium.TileLayer('OpenStreetMap').add_to(hcm_map)
folium.LayerControl().add_to(hcm_map)

hcm_map
```

## 2. Add Marker

```python
df_hcm = pd.read_excel('data/HCMC_location.xlsx', sheet_name="Location", engine = 'openpyxl')
df_hcm = df_hcm.dropna(subset=['Latitude', 'Longitude'])
df_hcm.head()
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=13)

for name, lat, lng, pop in zip(df_hcm.Name, df_hcm.Latitude, df_hcm.Longitude, df_hcm.Population):
    folium.CircleMarker(
        location=[lat, lng], 
        radius=pop/30,
        popup=name, 
        tooltip=name,
        color='crimson', 
        opacity=0.6, 
        fill=True,
        fill_color='red').add_to(hcm_map)

hcm_map
```
