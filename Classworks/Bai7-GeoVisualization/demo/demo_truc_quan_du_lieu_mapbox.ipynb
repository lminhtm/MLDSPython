```python
fig = px.scatter_mapbox(data_frame=df, lat='lat', lon='lon',
                       hover_name='City', hover_data=['State', 'Population'],
                       color_discrete_sequence=['deeppink'],
                       zoom=3, height=300)
fig.update_layout(mapbox_style='open-street-map', margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
import json
```

```python
f = open('../data/us-states.json')
states = json.load(f) 
f.close()
```

```python
states['features'][0]
```

```python
df = pd.read_csv('../data/US_Unemployment_Oct2012.csv')
```

```python
df.head()
```

```python
fig = px.choropleth(df, geojson=states, locations='State', color='Unemployment',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           scope="usa",
                           labels={'Unemployment':'unemployment rate'}
                          )

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
def get_name(id):
    for i in range(len(states['features'])):
        if states['features'][i]['id']==id:
            return states['features'][i]['properties']['name']
```

```python
for i in range(df.shape[0]):
    state = df.loc[i, 'State']    
    df.loc[i, 'Name'] = get_name(state)
```

```python
df.head()
```

```python
fig = px.choropleth(df, geojson=states, locations='State', color='Unemployment',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           scope="usa",
                           hover_name='Name',
                           labels={'Unemployment':'unemployment rate'}
                          )

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
