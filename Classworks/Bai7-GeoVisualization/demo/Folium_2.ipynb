```python
import pandas as pd
import folium
```

```python
df_hcm = pd.read_excel('../data/HCMC_location.xlsx', sheet_name="Location", engine = 'openpyxl')
df_hcm.head()
```

```python
hcm_latitude = 10.762622 
hcm_longitude = 106.660172
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
hcm_map
```

```python
df_hcm = df_hcm.dropna(subset=['Latitude', 'Longitude'])
df_hcm.head()
```

### Tạo các Circle Marker

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=13)

for lat, lng in zip(df_hcm.Latitude, df_hcm.Longitude):
    folium.CircleMarker(
        location=[lat, lng], 
        radius=10, color='yellow', 
        fill_opacity=0.6, 
        fill=True,
        fill_color='red').add_to(hcm_map)

hcm_map
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat, lng, pop in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Population):
    folium.CircleMarker(location=[lat, lng], 
                        radius=pop/50, color='crimson', 
                        opacity=0.6, fill=True,).add_to(hcm_map)

hcm_map
```

### Tạo các Popup

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat, lng, name in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Name):
    folium.Marker([lat, lng], popup=name).add_to(hcm_map)    #tooltip = name

hcm_map
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat, lng, name, pop in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Name, df_hcm.Population):
    folium.CircleMarker(location=[lat, lng], 
                        radius=pop/50, color='crimson', 
                        opacity=0.6, fill=True, popup=name).add_to(hcm_map)

hcm_map
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat, lng, name, pop in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Name, df_hcm.Population):
    tt = 'Dân số: '+str(pop)
    folium.Marker([lat, lng], popup=name, tooltip=tt).add_to(hcm_map)    
    
hcm_map
```

```python
def set_color(pop):
    if pop<=200:
        return 'blue'
    elif pop<=500:
        return 'green'
    else:
        return 'red'
```

```python
df_hcm['color'] = df_hcm.Population.map(lambda pop: set_color(pop))
```

```python
df_hcm.head()
```

```python
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat, lng, name, pop, color in zip(df_hcm.Latitude, df_hcm.Longitude, df_hcm.Name, df_hcm.Population, df_hcm.color):
    tt = name+' - Dân số: '+str(pop)
    folium.CircleMarker([lat, lng], tooltip=tt, color=color, radius=pop/30, opacity=0.6, fill=True).add_to(hcm_map) 
    
hcm_map
```
